# Fun Project: Password Generator 🔐
This is a simple password generator project.
Tried to make it easy and based on my understanding.

## Idea
This program generates a random password using letters, numbers and symbols.
It can be useful for creating strong passwords.

## Importing required libraries

```python
import random
import string
```

## Simple Password Generator

```python
length = int(input('Enter password length: '))

characters = string.ascii_letters + string.digits + string.punctuation

password = ''
for i in range(length):
    password += random.choice(characters)

print('Generated Password:', password)
```

## Improved Version

```python
length = int(input('Enter length: '))

all_chars = string.ascii_lowercase + string.ascii_uppercase + string.digits + string.punctuation

password = ''.join(random.choice(all_chars) for i in range(length))

print('Password:', password)
```
