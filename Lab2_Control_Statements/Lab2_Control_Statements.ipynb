## What are Control Statements?
Control statements help Python decide **what to do next** based on conditions.
Like if something is true → do this, else → do something else.

These are very important in programming.

## 1. If-Else Statement

```python
# Example 1
num = 5

if num > 0:
    print('Number is positive')
else:
    print('Number is negative')
```

```python
# example 2 (with input)
num = int(input('Enter a number: '))

if num % 2 == 0:
    print('Even number')
else:
    print('Odd number')
```

## 2. If-Elif-Else

```python
marks = int(input('Enter marks: '))

if marks >= 90:
    print('Grade A')
elif marks >= 70:
    print('Grade B')
elif marks >= 50:
    print('Grade C')
else:
    print('Fail')
```

## 3. Nested If

```python
age = int(input('Enter age: '))

if age >= 18:
    print('Adult')
    if age >= 60:
        print('Senior citizen')
else:
    print('Minor')
```

## 4. Loops
Loops are used when we want to repeat something multiple times.
(Mainly for and while)

### For Loop

```python
# printing numbers 0 to 4
for i in range(5):
    print(i)
```

```python
# printing list
fruits = ['apple', 'banana', 'mango']

for f in fruits:
    print(f)
```

### While Loop

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

## 5. break and continue

```python
# break example
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# continue example
for i in range(5):
    if i == 2:
        continue
    print(i)
```

## 6. Practice

```python
# simple program to print sum of numbers from 1 to n
n = int(input('Enter n: '))

sum = 0
for i in range(1, n+1):
    sum += i

print('Sum is:', sum)
```
