## What is NumPy?
NumPy is a Python library used for numerical operations.
It is faster than lists and useful for arrays and mathematical work.

## Importing NumPy

```python
import numpy as np

print('NumPy imported')
```

## Creating Arrays

```python
# 1D array
arr1 = np.array([1, 2, 3, 4])
print(arr1)
```

```python
# 2D array
arr2 = np.array([[1,2,3],[4,5,6]])
print(arr2)
```

## Array Properties

```python
print(arr2.shape)   # rows and columns
print(arr2.ndim)    # dimension
print(arr2.size)    # total elements
```

## Special Arrays

## Using arange()

```python
x = np.arange(0,10,2)
print(x)
```

## Basic Operations

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
print(a * b)
```

## Indexing

```python
arr = np.array([10,20,30,40])
print(arr[0])
print(arr[2])
```

## Reshape

```python
arr = np.array([1,2,3,4,5,6])
new_arr = arr.reshape(2,3)
print(new_arr)
```
