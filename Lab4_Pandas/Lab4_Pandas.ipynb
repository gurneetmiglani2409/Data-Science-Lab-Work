## What is Pandas?
Pandas is a Python library used for working with data (like tables).
It is very useful for data analysis and works similar to Excel in some ways.

## Importing Pandas

```python
import pandas as pd

print('Pandas imported')
```

## Pandas Series

```python
data = [10, 20, 30, 40]
s = pd.Series(data)
print(s)
```

## DataFrame

```python
data = {
    'Name': ['Aman', 'Riya', 'John'],
    'Age': [20, 21, 22]
}

df = pd.DataFrame(data)
print(df)
```

## Basic Functions

```python
print(df.head())
print(df.tail())
print(df.shape)
print(df.columns)
```

## Selecting Data

```python
print(df['Name'])
print(df[['Name', 'Age']])
```

## Filtering Data

```python
print(df[df['Age'] > 20])
```

## Adding New Column

```python
df['Marks'] = [85, 90, 88]
print(df)
```
